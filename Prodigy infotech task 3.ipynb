```python
column = df[['campaign','duration']]
q1 = np.percentile(column, 25)
q3 = np.percentile(column, 75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
df[['campaign','duration']] = column[(column > lower_bound) & (column < upper_bound)]
```

```python
df.plot(kind='box', subplots=True, layout=(2,5), figsize=(20,10), color='#808000')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=[np.number])
corr = numeric_df.corr()
print(corr)
corr = corr[abs(corr)>=0.90]
sns.heatmap(corr,annot=True,cmap='Set3',linewidth=0.2)
plt.show()
```
